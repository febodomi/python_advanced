```python
url='https://futar.bkk.hu/api/query/v1/ws/otp/api/where/schedule-for-stop.json?'
paramteres={'key':'apaiary-test'
            ,'version':'3'
            ,'appVersion':'apiary-1.0'
            ,'includeReferences':'true'
            ,'stopId':'BKK_F03557'
            ,'onlyDepartures':'true'
            ,'date':'20210324'}
```

```python
search=requests.get(url, params=paramteres)
print(search)
print(search.text)
```

```python
data=json.loads(search.text)
data['data']['entry']['schedules'][0]['directions'][0]['groups']
```

```python
data['data']['entry']['schedules'][0]['directions'][0]['stopTimes']
```

```python
data=json.loads(search.text)

for i in data['data']['entry']['schedules'][0]['directions'][0]['stopTimes']:
    print(i['stopHeadsign'],'-felé, indulás: ', datetime.fromtimestamp(i['departureTime']))
```

```python
data=json.loads(search.text)

for i in data['data']['entry']['schedules'][0]['directions'][0]['stopTimes']:
    if 'predictedDepartureTime' in i.keys():
        print(i['stopHeadsign'],'-felé, indulás: ', datetime.fromtimestamp(i['departureTime']), 'predicted:',datetime.fromtimestamp(i['predictedDepartureTime']))
    else:
        print(i['stopHeadsign'],'-felé, indulás: ', datetime.fromtimestamp(i['departureTime']))
```

```python
headsign=[]
departure=[]
prediction=[]
late=[]

for i in data['data']['entry']['schedules'][0]['directions'][0]['stopTimes']:
    if 'stopHeadsign' in i.keys():
        headsign.append(i['stopHeadsign'])
    else:
        headsign.append('No stopHeadsign')
        
    
    if 'departureTime' in i.keys():
        departure.append(datetime.fromtimestamp(i['departureTime']))
    else:
        departure.append('No departureTime')
        
    
    if 'predictedDepartureTime' in i.keys():
        prediction.append(datetime.fromtimestamp(i['predictedDepartureTime']))
    else:
        prediction.append('No predictedDepartureTime')
        
    if (('predictedDepartureTime' in i.keys()) and ('departureTime' in i.keys())):
        late.append(abs(datetime.fromtimestamp(i['predictedDepartureTime'])-datetime.fromtimestamp(i['departureTime'])))
    else:
        late.append('N/A')
```

```python
df = pd.DataFrame(list(zip(headsign, departure, prediction, late)),columns =['stopHeadsign', 'departureTime', 'predictedDepartureTime', 'late'])
df
```
